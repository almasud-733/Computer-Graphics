### Mid Point Line drawing: Calculate mid point lines between starting point(9,18) and ending point(14,22)by using Mid Point Line Drawing Algorithm

```python
import matplotlib.pyplot as plt
```

```python
def midpoint(x1, y1, x2, y2): 
    points = [] 
 
    dx = x2 - x1 
    dy = y2 - y1 
 
    x, y = x1, y1 
    points.append((x, y)) 
 
    d = dy - (dx / 2) 
 
    while x < x2: 
        x += 1 
        if d < 0: 
            d = d + dy 
        else: 
            y += 1 
            d = d + dy - dx 
        points.append((x, y)) 
 
    return points 
x1, y1 = 9,18
x2, y2 = 14,22
line_points = midpoint(x1, y1, x2, y2)
x_coords = [p[0] for p in line_points]
y_coords = [p[1] for p in line_points]

for (x1, y1) in line_points: 
    plt.text(x1, y1, f'({x1},{y1})', fontsize=9, ha='right', color='black', va='bottom')
plt.plot(x_coords, y_coords, marker="*", color="red")
plt.title("Mid-point Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
