#### In computer graphics, transformation is the process of changing the position, size, orientation, or shape of objects in a coordinate system using mathematical operations. It is essential for moving, resizing, rotating, and altering shapes in 2D and 3D graphics.

### Five common types are:

Translation – Moves an object from one place to another without changing its shape or size.

Scaling – Changes the size of an object, either enlarging or shrinking it.

Rotation – Turns an object around a fixed point at a certain angle.

Reflection – Flips an object to produce a mirror image over a specified axis or plane.

Shearing – Distorts the shape of an object by shifting its sides, creating a slanted effect.



### 1.Translation

```python
import matplotlib.pyplot as plt

square = [(0,3),(3,3),(3,0),(0,0)]
tx, ty = 1,1

translated = [(x + tx, y + ty) for x, y in square]

square += [square[0]]
translated += [translated[0]]

for (x, y) in translated:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='green')
for (x, y) in square:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='red')

plt.plot(*zip(*square), label='Original', color='blue')
plt.plot(*zip(*translated), '--', label='Translated', color='green')
plt.title('Square Translation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 2. Scaling

```python
import matplotlib.pyplot as plt

square =  [(0,3),(3,3),(3,0),(0,0)]
sx, sy = 2, 3

scaled = [(x * sx, y * sy) for x, y in square]

square += [square[0]]
scaled += [scaled[0]]

for (x, y) in scaled:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='green')
for (x, y) in square:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='red')

plt.plot(*zip(*square), label='Original', color='blue')
plt.plot(*zip(*scaled), '--', label='Scaled', color='orange')

plt.title('Square Scaling')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 3.Rotation

```python
import matplotlib.pyplot as plt
import math

triangle = [(0,0), (1,0), (1, 1)]
angle = 90
theta = math.radians(angle)

rotated = [
    (round(x * math.cos(theta) - y * math.sin(theta), 2),
     round(x * math.sin(theta) + y * math.cos(theta), 2))
    for x, y in triangle
]

triangle += [triangle[0]]
rotated += [rotated[0]]

for (x, y) in rotated:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='green')
for (x, y) in triangle:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=9, color='red')

plt.plot(*zip(*triangle), label='Original', color='blue')
plt.plot(*zip(*rotated), '--', label=f'Rotated ({angle}°)', color='red')

plt.title('Triangle Rotation')
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 3. Reflection

```python
import matplotlib.pyplot as plt

triangle = [(3, 4), (6, 4), (5, 6)]

def plot_polygon(points, *args, **kwargs):
    x, y = zip(*points)
    plt.plot(x + (x[0],), y + (y[0],), *args, **kwargs)


# Plot all triangles
plot_polygon(triangle, label='Original', color='blue')
plot_polygon([(x, -y) for x, y in triangle], '--', label='Reflected over X-axis', color='green')
plot_polygon([(-x, y) for x, y in triangle], ':', label='Reflected over Y-axis', color='red')

# Add labels and formatting
plt.title('Triangle Reflection')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 5. Shearing

```python
import matplotlib.pyplot as plt

triangle = [(1, 1), (0, 0), (1, 0)]
shear_x, shear_y = 2,3

def plot_polygon(points, *args, **kwargs):
    x, y = zip(*points)
    plt.plot(x + (x[0],), y + (y[0],), *args, **kwargs)

plot_polygon(triangle, label='Original', color='blue')
plot_polygon([(x + shear_x * y, y) for x, y in triangle],
             '--', label='Sheared in X', color='purple')
plot_polygon([(x, y + shear_y * x) for x, y in triangle],
             ':', label='Sheared in Y', color='orange')

plt.title('Triangle Shearing')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
