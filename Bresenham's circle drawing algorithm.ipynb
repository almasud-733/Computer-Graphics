### Given center point coordinates (xc, yc) = (10, 10) and radius r =8, generate all the points to form a circle by using Bresenhamâ€™s Circle drawing algorithm.

```python
import matplotlib.pyplot as plt
```

```python
def bresenham_circle(xc, yc, r):
    points = []
    
    x = 0
    y = r
    d = 3 - 2 * r  

    while x <= y:   
        points.extend([
            ( xc + x, yc + y),
            ( xc - x, yc + y),
            ( xc + x, yc - y),
            ( xc - x, yc - y),
            ( xc + y, yc + x),
            ( xc - y, yc + x),
            ( xc + y, yc - x),
            ( xc - y, yc - x)
        ])

        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        x += 1
    return points
```

```python
center_x, center_y = 0, 0
radius = 8
circle_points = bresenham_circle(center_x, center_y, radius)
x_coords = [p[0] for p in circle_points]
y_coords = [p[1] for p in circle_points]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, color='blue', marker='o')
plt.scatter(center_x, center_y, color='red', label='Center')

for (x, y) in circle_points:
    plt.text(x + 0.1, y + 0.1, f"({x},{y})", fontsize=8, color='black')

plt.title("Bresenham's Circle Drawing")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```
