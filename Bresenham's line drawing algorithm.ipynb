### Calculate the points between the starting point(9,18) and ending point(14,22) using Bresenham's line drawing algorithm.

```python
import matplotlib.pyplot as plt 
```

```python

 
def Bresenham(x1, y1, x2, y2): 
    points = [] 
    dx = x2 - x1 
    dy = y2 - y1 
 
    x, y = x1, y1 
    points.append((x, y)) 
 
    p = 2 * dy - dx 
 
    for _ in range(dx): 
        x += 1 
        if p < 0: 
            p = p + 2 * dy 
        else: 
            p = p + 2 * dy - 2 * dx 
            y += 1 
        points.append((x, y)) 
 
    return points
 
x1, y1 = 9,18
x2, y2 = 14,22
line_points = Bresenham(x1, y1, x2, y2)

x_coords = [p[0] for p in line_points]
y_coords = [p[1] for p in line_points]

for (x1, y1) in line_points: 
    plt.text(x1, y1, f'({x1},{y1})', fontsize=9, ha='right', color='black', va='bottom')
plt.plot(x_coords, y_coords, marker="o", color="blue")
plt.title("Bresenham's Line Drawing")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
