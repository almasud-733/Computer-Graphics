#### Calculated the intermediate points of a line using the DDA (Digital Differential Analyzer) line drawing algorithm. 
Given: 
â€¢ Starting point: (1,1) and Ending point:   (4, 6)  

```python
import matplotlib.pyplot as plt
```

```python
def DDA(x1, y1, x2, y2):
    points = []  
    dx = x2 - x1
    dy = y2 - y1

    steps = max(abs(dx), abs(dy))
    Xinc = dx / steps
    Yinc = dy / steps

    x, y = x1, y1
    for i in range(steps + 1):  
        points.append((round(x), round(y))) 
        x += Xinc
        y += Yinc
    return points

x1, y1 = 1, 1
x2, y2 = 4, 3
line_points = DDA(x1, y1, x2, y2)

print("DDA Line Points:")
for p in line_points:
    print(p)

x_coords = [p[0] for p in line_points]
y_coords = [p[1] for p in line_points]

for (x1, y1) in line_points: 
    plt.text(x1, y1, f'({x1},{y1})', fontsize=9, ha='right', color='red', va='bottom')
plt.plot(x_coords, y_coords, marker="o", color="blue")
plt.title("DDA Line Drawing Algorithm")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
